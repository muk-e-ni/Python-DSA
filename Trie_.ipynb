```python
class Node:
  def __init__(self):
    self.children = dict()
    self.is_end_of_word = False

class Trie:
  def __init__(self):
    self.root = Node()

  #O(m) Where m is the length of the trie
  def insert(self, word):
    current_word = self.root

    for char in word:
      if char not in current_word.children:
        current_word.children[char] = Node()
      current_word = current_word.children[char]

    current_word.is_end_of_word = True

  #O(m) Where m is the length of the trie
  def search(self, word):
    current_word = self.root
    for char in word:
      if char not in current_word.children:
        return False
      current_word = current_word.children[char]

    return current_word.is_end_of_word

  #O(m) Where m is the length of the trie
  def has_prefix(self, prefix):
    current_word = self.root
    for char in prefix:
      if char not in current_word.children:
        return False
      current_word = current_word.children[char]

    return True

  #O(m + k) Where m is the prefix length and k the total number of characters in all suffixes
  def starts_with(self, prefix):
    current_node = self.root
    words = []

    for char in prefix:
      if char not in current_node.children:
        return words

      current_node = current_node.children[char]

    def _dfs(current_node, path):
        if current_node.is_end_of_word:
          words.append(''.join(path))

        for c, child_node in current_node.children.items():
          _dfs(child_node, path + [c])

    _dfs(current_node, list(prefix))
    return words

   #O(n) where n is the number of nodes in the trie
  def list_words(self):
    words = []

    def _dfs(node, path):
        if node.is_end_of_word:
          words.append(''.join(path))

        for c, child_node in node.children.items():
          _dfs(child_node, path + [c])

    _dfs(self.root, [])

    return words

  #O(m) Where m is the length of the trie
  def delete(self, word):
    self._delete(self.root, word, 0)

  def _delete(self, current_node, word, index: int):
    if index == len(word):
      if not current_node.is_end_of_word:
        return False
      current_node.is_end_of_word = False

      return len(current_node.children) == 0

    char = word[index]
    char_node = current_node.children.get(char)

    if char_node is None:
      return False

    should_delete_current_node = self._delete(char_node, word, index+1)
    if should_delete_current_node:
      del current_node.children[char]
      return len(current_node.children) == 0 and not current_node.is_end_of_word

    return False
```

```python
if __name__ == '__main__':
  trie = Trie()

  trie.insert('apple')
  trie.insert('app')
  trie.insert('banana')
  trie.insert('bat')
  trie.insert('bar')
  trie.insert('carriage')
  trie.insert('cat')
  trie.insert('car')
  trie.insert('contains')
  trie.insert('carton')
  trie.insert('cats')

  print("Original list")

  print(trie.list_words())

  print("Searching for 'app'")

  print(trie.search('app'))

  print("Searching for prefix 'app'")

  print(trie.has_prefix('app'))

  print("Searching for prefix 'car'")

  print(trie.has_prefix('car'))
  print ("deleting  app from the list")

  trie.delete('app')

  print("List after deletion")

  print(trie.list_words())

  print("Searching for all words that start with 'car'")

  print(trie.starts_with('ca'))
```
