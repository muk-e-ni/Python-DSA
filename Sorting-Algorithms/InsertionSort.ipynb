```python
def insertion_sort(array):

  n = len(array)
  for i in range(1, n):
    current_element = array[i]
    previous_element_from_current = i -1

    while previous_element_from_current >=0 and array[previous_element_from_current] > current_element:
      array[previous_element_from_current + 1] = array[previous_element_from_current]

      previous_element_from_current -= 1

    array[previous_element_from_current + 1 ] = current_element

  return array

print("Insertion with the While Loop\n")
my_array = [64, 34, 25, 12, 22, 11, 90, 5, 756, 57, 546]
print("Unsorted Array \n", my_array, end='\n\n')

sorted_array = insertion_sort(my_array)

print("Sorted Array \n", sorted_array)



# using the nested for loop

def insertion_sort_nested_if(array):

  n = len(array)
  for i in range(1, n):
    for j in range(i, 0, -1):
      if array[j] < array[j-1]:
        array[j], array[j-1] = array[j-1], array[j]
      else:
        break

  return array

print("\nInsertion with the Nested If \n")

my_array2 = [3, 4, 56,2423,-2,13, 2343, 34, 10, 200, 24, 45]
print("Unsorted Array \n", my_array2, end='\n\n')

sorted_array2 = insertion_sort_nested_if(my_array2)
print("Sorted with Nested if statements \n", sorted_array2)
```
