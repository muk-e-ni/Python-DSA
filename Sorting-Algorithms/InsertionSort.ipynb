```python
def insertion_sort(array):

  n = len(array)
  for i in range(1, n):
    current_element = array[i]
    previous_element_from_current = i -1

    while previous_element_from_current >=0 and array[previous_element_from_current] > current_element:
      array[previous_element_from_current + 1] = array[previous_element_from_current]

      previous_element_from_current -= 1

    array[previous_element_from_current + 1 ] = current_element

  return array\

my_array = [64, 34, 25, 12, 22, 11, 90, 5, 756, 57, 546]
print("Unsorted Array \n", my_array, end='\n\n')

sorted_array = insertion_sort(my_array)

print("Sorted Array \n", sorted_array)
```
