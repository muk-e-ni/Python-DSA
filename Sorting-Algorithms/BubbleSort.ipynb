```python
A = [-5,3,2,1,-3,-4,6,10,100,45]
def bubble_sort(array):

  n = len(array)
  for i in range(n-1):
    for j in range(n-i-1):
      if array[j] > array[j+1]:
        array[j], array[j+1] = array[j+1], array[j]
print("Unsorted array\n", A)
bubble_sort(A)
print("Bubble Sorted array\n",A)

```
