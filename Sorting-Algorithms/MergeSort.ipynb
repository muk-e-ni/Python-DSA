```python
############# MERGE SORT ALGORITHM ############3

# Time complexity O(n * log n)
# Space complexity: O(n)

def Merge_sort(array):
  n = len(array)

  if n == 1:
    return array

  m = n // 2
  left = array[:m]
  right = array[m:]

  left_array = Merge_sort(left)
  right_array = Merge_sort(right)

  L_len = len(left_array)
  R_len = len(right_array)

  l = r = 0

  sorted_array = [0] * n

  i = 0

  while l < L_len and r < R_len:
    if left_array[l] < right_array[r]:
      sorted_array[i] = left_array[l]
      l += 1
      i += 1
    else:
      sorted_array[i] = right_array[r]
      r += 1
      i += 1

  while l < L_len:
    sorted_array[i] = left_array[l]
    l+=1
    i+=1

  while r < R_len:
    sorted_array[i] = right_array[r]
    r+=1
    i+=1

  return sorted_array

my_array = [1, 2, -2, 23, 24, 56, 34, 3, 64, 35, 34, 54, 4, 24, 64, 64]

print("Unsorted Array\n", my_array, end="\n\n")
print("Merge sorted Array\n", Merge_sort(my_array))





```
