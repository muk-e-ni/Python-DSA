```python
################# COUNTING SORT ALGORITHM ############
# Time complexity: O(k + n) where n is the maximum number in the array
# Space complexity: O(n)
# This fucntion only works with positive numbers

def Quick_sort(array):

  maxx = max(array) + 1

  counts_array = [0] * maxx

  for i in array:
    counts_array[i] +=1

  i = 0

  for c in range(maxx):
    while counts_array[c] > 0:
      array[i] = c
      i += 1
      counts_array[c] -= 1

  return array

array = [1, 4, 1, 69, 7, 5, 2, 45, 64, 454, 23, 54, 34]

print("Unsorted array\n", array, end='\n\n')
print("Counting Sorted array\n", Quick_sort(array))


```
