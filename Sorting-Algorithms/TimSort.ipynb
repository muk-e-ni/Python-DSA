```python
############# TIM SORT ##########

# buit-in python method

# Time complexity: O(n log n)

array = [4, -4, 5, 45, 5, 23, -4, 56]

# sorting in constant space
array.sort()

print(array)
```

```python
# sorting items in a list of tuples

my_list = [(-3, 4), (34, -4), (5,76), (5,6), (45, 4), (9,5), (7, 7)]

#sorting based on the first position of the tuples
sorted_list = sorted(my_list, key = lambda t: t[0])

print("Sorted based on the first position of the tuples in the list\n", sorted_list, end='\n\n')

sorted_list2 = sorted(my_list, key = lambda t: t[1])

print("Sorted based on the second position of the tuples in the list\n", sorted_list2, end='\n\n')

sorted_list3 = sorted(my_list, key = lambda t: -t[0])
print("sorted list in reverse order of the first position\n", sorted_list3)
```
