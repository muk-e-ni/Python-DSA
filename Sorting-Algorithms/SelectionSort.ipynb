```python
############SELECTION SORT############

# Time complexity O(n^2)
# Space Complexity O(1)

def Selection_sort(array):
  n = len(array)
  for i in range(n):
    min_index = i
    for j in range(i+1, n):
      if array[j] < array[min_index]:
        min_index = j
    array[i], array[min_index] = array[min_index], array[i]

  return array

my_array = [-3, 5, 6, 78, -5, 9, 78, 100, 90, 24, 56 , -9]

print("Unsorted array\n", my_array, end="\n\n")

print("Selection sorted array\n", Selection_sort(my_array))
```
