```python
from collections import deque # to reduce time complexity of the queue
import numpy as np

class Graph:
  def __init__(self, directed = False):
    self.directed = directed
    self.adj_list = dict()

  def __repr__(self):
    graph_string = ""
    for v, adj in self.adj_list.items():
      graph_string += f"{v}:> {adj}\n"
    return graph_string

  def add_vertex(self, vertex):
    if vertex not in self.adj_list:
      self.adj_list[vertex] = set()

    else:
      raise ValueError(f"Vertex {vertex} already in graph")

  def remove_vertex(self, vertex):
    if vertex not in self.adj_list:
      raise ValueError(f"Vertex {vertex} not in graph")

    else:
      for neighbours in self.adj_list.values():
        if vertex in neighbours:
          neighbours.remove(vertex)

      del self.adj_list[vertex]
  def add_edge(self, v1, v2, weight=None):
    if v1 not in self.adj_list:
      self.add_vertex(v1)

    if v2 not in self.adj_list:
      self.add_vertex(v2)

    if weight is None:
      self.adj_list[v1].add9(v2)
      if not self.directed:
        self.adj_list[v2].add(v1)

    else:
      self.adj_list[v1].add((v2, weight))
      if not self.directed:
        self.adj_list[v2].add((v1, weight))
  def remove_edge(self,v1, v2):
    if v1 not in self.adj_list or v2 not in self.adj_list:
      raise ValueError('Edge does not exist')

    for neighbour in list(self.adj_list[v1]):
      if isinstance(neighbour, tuple):
        if neighbour[0] == v2:
          self.adj_list[v1].remove(neighbour)
          break
      else:
        if neighbour == v2:
          self.adj_list[v1].remove(neighbour)
    if not self.directed:
      for neighbour in self.adj_list[v2]:
        if isinstance(neighbour, tuple):
          if neighbour[0] == v1:
           self.adj_list[v2].remove(neighbour)
           break

        else:
          if neighbour == v1:
            self.adj_list[v2].remove(neighbour)
    else:
      raise ValueError('Edge does not exist')

  def get_neighbours(self, v):
    if v not in self.adj_list:
      raise ValueError(f"Vertex {v} not in graph")

    else:
      return self.adj_list.get(v, set())
  def get_vertexes(self):
    return list(self.adj_list.keys())
  def has_vertex(self, v):
    return v in self.adj_list

  def has_edge(self, v1, v2):
    if v1 not in self.adj_list:
      return False

    for neighbour in self.adj_list[v1]:
      if isinstance(v2, tuple):
        if neighbour[0] == v2:
          return True
      else:
        if neighbour[0] == v2:
          return True
      return False

  def get_edges(self):
    edges = []
    for v1, neighbours in self.adj_list.items():
      for v2 in neighbours:
        edges.append((v1, v2))
    return edges

  # uses queue data structure

  def bfs(self, start):
    visited_vertexes = set()
    queue = deque([start])
    order = []

    while queue:
      vertex = queue.popleft()
      if vertex not in visited_vertexes:
        visited_vertexes.add(vertex)
        order.append(vertex)
        neighbours = self.get_neighbours(vertex)
        for neighbour in neighbours:
          if isinstance(neighbour, tuple):
            neighbour = neighbour[0]
          if neighbour not in visited_vertexes:
              queue.append(neighbour)
    return order

  # uses stack data structure
  def dfs(self, start):
    visited_vertexes = set()
    stack = [start]
    order = []

    while stack:
      vertex = stack.pop()
      if vertex not in visited_vertexes:
        visited_vertexes.add(vertex)
        order.append(vertex)
        neighbours = self.get_neighbours(vertex)
        for neighbour in neighbours:
          if isinstance(neighbour, tuple):
            neighbour = neighbour[0]
          if neighbour not in visited_vertexes:
            stack.append(neighbour)
    return order
  def to_adj_matrix(self):
    vertexes = self.get_vertexes()
    index = {vertex: i for i, vertex in enumerate(vertexes)}
    matrix = [[0 for _ in range(len(vertexes))] for _ in range(len(vertexes))]
    for v1, neighbours in self.adj_list.items():
      for v2 in neighbours:
        if isinstance(v2, tuple):
          to, weight = v2
          matrix[index[v1]][index[to]] = weight
        else:
          matrix[index[v1]][index[v2]] = 1
    return matrix, vertexes



```
