```python
class MinHeap:
  def __init__(self):
    self.heap = []

  def __len__(self):
    return len(self.heap)

  def __repr__(self):
    return str(self.heap)

  # O(log n) time complexity
  def insert(self, key, val):
    self.heap.append([key, val])
    self._sift_up(len(self.heap) - 1)

  #O(1) time complexity
  def peek(self):
    if not self.heap:
      raise IndexError("Heap is empty")
    return self.heap[0]

  #O(log n) time complexity
  def extract_min(self):
    if not self.heap:
      raise IndexError("Heap is empty")
    min_element = self.heap[0]
    last_element = self.heap.pop()

    if self.heap:
      self.heap[0] = last_element
      self._sift_down(0)
    return min_element

  # O(n) time complexity
  def heapify(self, elements):

    self.heap = list(elements)

    for i in reversed(range(self._parent(len(self.heap) -1 ) + 1)):
      self._sift_down(i)

  def meld(self, other):
    combined_heap = self.heap + other.heap
    self.heapify(combined_heap)

    other.heap = []

  # O(1) time complexity
  def _parent(self, index):
    return (index - 1)//2 if index !=0 else None

  # O(1) time complexity
  def _left(self, index):
    left = (2 * index) + 1
    return left if left < len(self.heap) else None

  # O(1) time complexity
  def _right(self, index):
    right = (2 * index) + 2
    return right if right < len(self.heap) else None

  # O(1) time complexity
  def _sift_up(self, index):
    parent_index = self._parent(index)

    while parent_index and self.heap[index][0] < self.heap[parent_index][0]:
      self._swap(index, parent_index)
      index = parent_index
      parent_index = self._parent(index)

  # O(log n) time complexity
  def _sift_down(self, index):
    while True:
      right_index = self._right(index)
      left_index = self._left(index)

      smallest_index = index

      if right_index is not None  and self.heap[right_index][0] < self.heap[smallest_index][0]:
        smallest_index = right_index

      if left_index is not None and self.heap[left_index][0] < self.heap[smallest_index][0]:
        smallest_index = left_index

      if smallest_index == index:
        break

      self.heap[index], self.heap[smallest_index] = self.heap[smallest_index], self.heap[index]

      index = smallest_index


 # O(1) time complexity
  def _swap(self, index, parent_index):
    self.heap[index], self.heap[parent_index] = self.heap[parent_index], self.heap[index]

```

```python
if __name__ == '__main__':
  heap = MinHeap()

  heap.heapify([[1,2],[2,2], [3,4], [4,4], [5,5]])
  print(heap)
```
